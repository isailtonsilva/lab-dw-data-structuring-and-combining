```python
# Step 0 - Import pandas for data loading and cleaning
import pandas as pd
import numpy as np
```

```python
# Step 0.1 - Define URLs for the 3 datasets
# These are the raw CSV files provided by the bootcamp

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
```

```python
# Step 0.2 - Load the datasets into DataFrames
# read_csv reads a CSV file into a pandas DataFrame

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

# Quick check
df1.head()
```

```python
# Step 1 - Standardize column names
# We make them lowercase and replace spaces with underscores for consistency

def clean_column_names(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
    )
    return df

df1 = clean_column_names(df1)
df2 = clean_column_names(df2)
df3 = clean_column_names(df3)
```

```python
# Step 2 - Clean invalid/inconsistent values
# This function applies fixes similar to the previous Data Cleaning lab

def clean_invalid_values(df):
    # Step 2.1 - Standardize gender values
    if "gender" in df.columns:
        df["gender"] = (
            df["gender"]
            .astype(str)
            .str.strip()
            .str.lower()
            .replace({
                "female": "F", "femal": "F", "f": "F",
                "male": "M", "m": "M"
            })
        )

    # Step 2.2 - Standardize state values (examples from lab instructions)
    if "state" in df.columns:
        df["state"] = df["state"].replace({
            "AZ": "Arizona",
            "Cali": "California",
            "WA": "Washington"
        })

    # Step 2.3 - Standardize education spelling
    if "education" in df.columns:
        df["education"] = df["education"].replace({
            "Bachelors": "Bachelor"
        })

    # Step 2.4 - Remove '%' from Customer Lifetime Value and convert to numeric
    if "customer_lifetime_value" in df.columns:
        df["customer_lifetime_value"] = (
            df["customer_lifetime_value"]
            .astype(str)
            .str.replace("%", "", regex=False)
        )
        df["customer_lifetime_value"] = pd.to_numeric(df["customer_lifetime_value"], errors="coerce")

    # Step 2.5 - Reduce vehicle class categories to simplify encoding
    if "vehicle_class" in df.columns:
        df["vehicle_class"] = df["vehicle_class"].replace({
            "Sports Car": "Luxury",
            "Luxury SUV": "Luxury",
            "Luxury Car": "Luxury"
        })

    return df

df1 = clean_invalid_values(df1)
df2 = clean_invalid_values(df2)
df3 = clean_invalid_values(df3)
```

```python
# Step 3 - Fix data types and formatting issues

def format_data_types(df):
    # Step 3.1 - Convert number_of_open_complaints to numeric
    # Some values look like '1/5/00' -> we take the middle number (index 1)
    if "number_of_open_complaints" in df.columns:
        df["number_of_open_complaints"] = (
            df["number_of_open_complaints"]
            .astype(str)
            .str.split("/")
            .str[1]
        )
        df["number_of_open_complaints"] = pd.to_numeric(df["number_of_open_complaints"], errors="coerce")

    # Step 3.2 - Convert effective_to_date to datetime if present
    if "effective_to_date" in df.columns:
        df["effective_to_date"] = pd.to_datetime(df["effective_to_date"], errors="coerce")

    return df

df1 = format_data_types(df1)
df2 = format_data_types(df2)
df3 = format_data_types(df3)
```

```python
# Step 4 - Handle missing values
# Numeric columns: fill with median
# Categorical columns: fill with mode

def handle_nulls(df):
    num_cols = df.select_dtypes(include=["int64", "float64"]).columns
    cat_cols = df.select_dtypes(include=["object"]).columns

    for col in num_cols:
        df[col] = df[col].fillna(df[col].median())

    for col in cat_cols:
        df[col] = df[col].fillna(df[col].mode()[0])

    return df

df1 = handle_nulls(df1)
df2 = handle_nulls(df2)
df3 = handle_nulls(df3)
```

```python
# Step 5 - Combine datasets using concat
# concat stacks the dataframes vertically (rows)
# ignore_index=True resets index for clean combined dataset

df_combined = pd.concat([df1, df2, df3], ignore_index=True)

# Quick checks
df_combined.shape, df_combined.head()
```

```python
# Step 6 - Check missing values and duplicates after combining

df_combined.isna().sum().sort_values(ascending=False).head(10)
```

```python
# Step 6.1 - Check duplicates and remove them if any

df_combined.duplicated().sum()
df_combined = df_combined.drop_duplicates()
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Step 7 - Load the cleaned marketing dataset
# This dataset is already cleaned according to the file name

url_marketing = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
marketing = pd.read_csv(url_marketing)

# Quick inspection
marketing.head()
```

```python
# Step 8 - Pivot table: total revenue (sum of CLV) by sales channel
# pivot_table groups the data and aggregates the values

pivot_revenue = marketing.pivot_table(
    index="sales_channel",
    values="customer_lifetime_value",
    aggfunc="sum"
).round(2)

pivot_revenue
```

```python
# Step 8.1 - Sort to see which channel is highest

pivot_revenue.sort_values(by="customer_lifetime_value", ascending=False)
```

```python
# Step 9 - Pivot table: average CLV by gender and education
# index and columns create a matrix-style summary

pivot_clv_gender_edu = marketing.pivot_table(
    index="gender",
    columns="education",
    values="customer_lifetime_value",
    aggfunc="mean"
).round(2)

pivot_clv_gender_edu
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Step 10 - Convert effective_to_date to datetime
# Extract month for grouping complaints by month

marketing["effective_to_date"] = pd.to_datetime(marketing["effective_to_date"], errors="coerce")
marketing["month"] = marketing["effective_to_date"].dt.month
```

```python
# Step 11 - Group and count complaints by policy type and month
# Long format means each row is one (policy_type, month) observation

complaints_long = (
    marketing
    .groupby(["policy_type", "month"])["number_of_open_complaints"]
    .sum()
    .reset_index()
    .sort_values(["policy_type", "month"])
)

complaints_long
```
